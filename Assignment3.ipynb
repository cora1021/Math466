```python
import csv
import os
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
currency_eurusd = 'EURUSD=X.csv'
eurusd_data = []
with open(currency_eurusd) as csvfile:
    csv_reader = csv.reader(csvfile)  
    currency_header = next(csv_reader)
    for row in csv_reader:  
        eurusd_data.append(row)
date = []
open_price = []
high = []
low = []
close = []
adj_close = []
volume = []
for i in range(len(eurusd_data)):
    date.append(eurusd_data[i][0])
    open_price.append(float(eurusd_data[i][1]))
    high.append(float(eurusd_data[i][2]))
    low.append(float(eurusd_data[i][3]))
    close.append(float(eurusd_data[i][4]))
    adj_close.append(float(eurusd_data[i][5]))
    volume.append(float(eurusd_data[i][6]))
```

```python
# Data Visualization
plt.title('Data Visualization')
plt.plot(open_price, color='green', label='open price')
plt.plot(high, color='red', label='high price')
plt.plot(low,  color='skyblue', label='low price')
plt.plot(close, color='blue', label='close price')
plt.legend() 
plt.xlabel('Date')
plt.show()
```

```python
open_price_piece_1 = np.array(open_price[9:22]) - open_price[9]
open_price_piece_2 = np.array(open_price[21:30]) - open_price[21]
open_price_piece_3 = np.array(open_price[29:36]) - open_price[29]
open_price_piece_4 = np.array(open_price[35:42]) - open_price[35]
open_price_piece_5 = np.array(open_price[69:98]) - open_price[69]
open_price_piece_6 = np.array(open_price[97:103]) - open_price[97]
open_price_piece_7 = np.array(open_price[102:125]) - open_price[102]
open_price_piece_8 = np.array(open_price[156:166]) - open_price[156]
open_price_piece_9 = np.array(open_price[190:239]) - open_price[190]
```

```python
x_range = [range(9, 22), range(21, 30), range(29, 36), range(35, 42), range(69, 98), range(97, 103), range(102, 125), range(156, 166), range(190, 239)]
```

```python
open_price_pieces = [open_price_piece_1,open_price_piece_2,open_price_piece_3,open_price_piece_4,open_price_piece_5,open_price_piece_6,\
                    open_price_piece_7,open_price_piece_8,open_price_piece_9]
```

```python
from scipy.optimize import curve_fit
import numpy as np
```

```python
def func(x, a, b, c, d):
    return a * np.power(x, 3) + b * np.power(x, 2) + c * x + d
```

```python
popt_1, pcov_1 = curve_fit(func, np.array((range(len(open_price_piece_1)))), np.array(open_price_piece_1))
popt_2, pcov_2 = curve_fit(func, np.array((range(len(open_price_piece_2)))), np.array(open_price_piece_2))
popt_3, pcov_3 = curve_fit(func, np.array((range(len(open_price_piece_3)))), np.array(open_price_piece_3))
popt_4, pcov_4 = curve_fit(func, np.array((range(len(open_price_piece_4)))), np.array(open_price_piece_4))
popt_5, pcov_5 = curve_fit(func, np.array((range(len(open_price_piece_5)))), np.array(open_price_piece_5))
popt_6, pcov_6 = curve_fit(func, np.array((range(len(open_price_piece_6)))), np.array(open_price_piece_6))
popt_7, pcov_7 = curve_fit(func, np.array((range(len(open_price_piece_7)))), np.array(open_price_piece_7))
popt_8, pcov_8 = curve_fit(func, np.array((range(len(open_price_piece_8)))), np.array(open_price_piece_8))
popt_9, pcov_9 = curve_fit(func, np.array((range(len(open_price_piece_9)))), np.array(open_price_piece_9))
```

```python
popts = [popt_1,popt_2,popt_3,popt_4,popt_5,popt_6,popt_7,popt_8,popt_9]
```

```python
from sklearn.cluster import KMeans
```

```python
def get_y_(i):
    a = popts[i][0]
    b = popts[i][1]
    c = popts[i][2]
    d = popts[i][3]
    y_ = func(np.array((range(len(open_price_pieces[i])))), a,b,c,d)
    return y_
```

```python
plt.plot(open_price)
for i in range(9):

    plt.plot(x_range[i], get_y_(i)+open_price[x_range[i][0]], 'r')
```

```python
def sign(x):
    
    if x>=0:
        return 1
    else:
        return -1
```

```python
k_feature_1 = np.array(list(popt_1) + [max(open_price_piece_1) - min(open_price_piece_1)] + [sign(popt_1[0])])
k_feature_2 = np.array(list(popt_2) + [max(open_price_piece_2) - min(open_price_piece_2)] + [sign(popt_2[0])])
k_feature_3 = np.array(list(popt_3) + [max(open_price_piece_3) - min(open_price_piece_3)] + [sign(popt_3[0])])
k_feature_4 = np.array(list(popt_4) + [max(open_price_piece_4) - min(open_price_piece_4)] + [sign(popt_4[0])])
k_feature_5 = np.array(list(popt_5) + [max(open_price_piece_5) - min(open_price_piece_5)] + [sign(popt_5[0])])
k_feature_6 = np.array(list(popt_6) + [max(open_price_piece_6) - min(open_price_piece_6)] + [sign(popt_6[0])])
k_feature_7 = np.array(list(popt_7) + [max(open_price_piece_7) - min(open_price_piece_7)] + [sign(popt_7[0])])
k_feature_8 = np.array(list(popt_8) + [max(open_price_piece_8) - min(open_price_piece_8)] + [sign(popt_8[0])])
k_feature_9 = np.array(list(popt_9) + [max(open_price_piece_9) - min(open_price_piece_9)] + [sign(popt_9[0])])
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(np.array([k_feature_1, k_feature_2, k_feature_3, k_feature_4, k_feature_5, k_feature_6, k_feature_7, k_feature_8, k_feature_9]))
```

```python
kmeans.labels_
```

```python
plt.plot(open_price)
for i in range(9):
    if kmeans.labels_[i]==0:
        color='r'
    else:
        color='g'
    plt.plot(x_range[i], get_y_(i)+open_price[x_range[i][0]], color)
```

```python
import csv
import os
import numpy as np
```

```python
currency_eurusd = 'EURUSD=X.csv'
```

```python
eurusd_data = []
with open(currency_eurusd) as csvfile:
    csv_reader = csv.reader(csvfile)  
    currency_header = next(csv_reader)
    for row in csv_reader:  
        eurusd_data.append(row)
```

```python
eurusd_data[0]
```

```python
date = []
open_price = []
high = []
low = []
close = []
adj_close = []
volume = []
for i in range(len(eurusd_data)):
    date.append(eurusd_data[i][0])
    open_price.append(float(eurusd_data[i][1]))
    high.append(float(eurusd_data[i][2]))
    low.append(float(eurusd_data[i][3]))
    close.append(float(eurusd_data[i][4]))
    adj_close.append(float(eurusd_data[i][5]))
    volume.append(float(eurusd_data[i][6]))
```

```python
import matplotlib.pyplot as plt
```

```python
# Data Visualization
plt.title('Data Visualization')
plt.plot(open_price, color='green', label='open price')
plt.plot(high, color='red', label='high price')
plt.plot(low,  color='skyblue', label='low price')
plt.plot(close, color='blue', label='close price')
plt.legend() 
plt.xlabel('Date')
plt.show()
```
